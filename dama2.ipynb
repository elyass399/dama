```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pygame
import time
from stable_baselines3 import PPO  # Usiamo PPO, spesso più veloce e stabile di DQN per questi problemi
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.env_checker import check_env
import os

# --- COSTANTI ---
VUOTO = 0
NERO_PEDINA = 1
NERO_DAMA = 2
BIANCO_PEDINA = 3
BIANCO_DAMA = 4
DIMENSIONE = 8
ACTION_DIM = DIMENSIONE**4

class CheckersEnvAdvanced(gym.Env):
    """
    Ambiente Dama Avanzato per Stable Baselines3.
    Regole aggiunte:
    1. Cattura Multipla Automatica.
    2. Regola Italiana: Le Pedine NON possono mangiare le Dame.
    """
    
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        # PPO lavora meglio con Discrete o MultiDiscrete. 
        # Manteniamo Discrete(4096) per compatibilità, ma usiamo Action Masking logico.
        self.action_space = spaces.Discrete(ACTION_DIM)
        
        # L'osservazione è la scacchiera 8x8
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()

    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}

    def _update_moves(self):
        """Calcola le mosse legali. Se ci sono catture, OBBLIGA a catturare."""
        self.possible_moves = self._find_all_legal_moves()

    def _decode_action(self, action_int):
        r1 = action_int // 512
        c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8
        c2 = action_int % 8
        return r1, c1, r2, c2

    def _find_all_legal_moves(self):
        captures = []
        normal = []
        king = self.player_map.get(self.current_player, 0)
        
        for r in range(8):
            for c in range(8):
                if self.board[r,c] == self.current_player or self.board[r,c] == king:
                    caps = self._get_captures(r, c)
                    captures.extend(caps)
                    if not captures: # Ottimizzazione: se ho già catture, non cerco mosse normali
                        normal.extend(self._get_moves(r, c))
        
        return captures if captures else normal

    def _get_moves(self, r, c):
        m = []
        piece = self.board[r, c]
        # Direzioni base
        dirs = [(1,-1),(1,1)] if piece == 1 else [(-1,-1),(-1,1)]
        if piece in [2, 4]: dirs = [(1,-1),(1,1),(-1,-1),(-1,1)]
        
        for dr, dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc] == 0:
                m.append((r,c,r+dr,c+dc))
        return m

    def _get_captures(self, r, c):
        caps = []
        piece = self.board[r, c]
        
        # Identifica nemici
        enemies = [3, 4] if piece in [1, 2] else [1, 2]
        
        # Direzioni
        dirs = [(1,-1),(1,1)] if piece == 1 else [(-1,-1),(-1,1)]
        if piece in [2, 4]: dirs = [(1,-1),(1,1),(-1,-1),(-1,1)]
        
        for dr, dc in dirs:
            rm, cm = r+dr, c+dc      # Casella in mezzo (nemico)
            rd, cd = r+2*dr, c+2*dc  # Casella destinazione
            
            if 0<=rd<8 and 0<=cd<8:
                mid_piece = self.board[rm, cm]
                if mid_piece in enemies and self.board[rd, cd] == 0:
                    
                    # --- NUOVA REGOLA: PEDINA NON MANGIA DAMA ---
                    # Se io sono pedina (1 o 3) e lui è Dama (2 o 4), NON POSSO MANGIARE.
                    if piece in [1, 3] and mid_piece in [2, 4]:
                        continue # Salta questa cattura
                        
                    caps.append((r,c,rd,cd))
        return caps

    def _execute_capture_chain(self, r, c):
        """Funzione ricorsiva o iterativa per mangiare finché possibile (Multicattura)."""
        chain_reward = 0
        just_captured = True
        
        curr_r, curr_c = r, c
        
        # Continua finché l'ultimo pezzo mosso può ancora mangiare
        # (Semplificazione Greedy: mangia il primo che trova nella catena)
        while just_captured:
            possible_caps = self._get_captures(curr_r, curr_c)
            if not possible_caps:
                break
            
            # Prende la prima cattura disponibile (Greedy)
            # In una versione pro, l'IA dovrebbe scegliere il percorso, 
            # ma per RL è meglio automatizzare la catena.
            _, _, next_r, next_c = possible_caps[0]
            
            # Esegui cattura
            mid_r, mid_c = (curr_r + next_r)//2, (curr_c + next_c)//2
            self.board[next_r, next_c] = self.board[curr_r, curr_c]
            self.board[curr_r, curr_c] = 0
            self.board[mid_r, mid_c] = 0
            
            chain_reward += 1.5 # Reward extra per ogni pezzo mangiato in catena
            curr_r, curr_c = next_r, next_c
            
        return curr_r, curr_c, chain_reward

    def step(self, action):
        r1, c1, r2, c2 = self._decode_action(action)
        move = (r1, c1, r2, c2)
        
        reward = 0
        terminated = False
        
        if move in self.possible_moves:
            # Esecuzione Mossa
            piece = self.board[r1, c1]
            
            # Caso CATTURA (Salto)
            if abs(r2 - r1) == 2:
                # Esegui il primo salto
                self.board[r2, c2] = piece
                self.board[r1, c1] = 0
                self.board[(r1+r2)//2, (c1+c2)//2] = 0
                reward += 1.0
                
                # --- NUOVA REGOLA: CATTURA MULTIPLA ---
                # Controlla se dal punto di arrivo (r2,c2) si può mangiare ancora
                # Se sì, il sistema lo fa in automatico (Chain)
                final_r, final_c, extra_rew = self._execute_capture_chain(r2, c2)
                reward += extra_rew
                r2, c2 = final_r, final_c # Aggiorna posizione finale
                
            else:
                # Mossa normale
                self.board[r2, c2] = piece
                self.board[r1, c1] = 0
            
            # --- PROMOZIONE DAMA ---
            # Se arriva in fondo, diventa Dama.
            if (self.board[r2, c2] == 1 and r2 == 7):
                self.board[r2, c2] = 2
                reward += 3.0
            elif (self.board[r2, c2] == 3 and r2 == 0):
                self.board[r2, c2] = 4
                reward += 3.0
            
            # Cambio Turno
            self.current_player = 3 if self.current_player == 1 else 1
            self._update_moves()
            
            # Controllo Fine Partita
            if not self.possible_moves:
                terminated = True
                reward += 10.0 # Vittoria
                
        else:
            # Mossa Illegale: Invece di crashare o loopare,
            # diamo penalità e NON cambiamo stato (ma l'episodio finisce per RL)
            # O meglio per SB3: Penalità forte e chiudi episodio.
            reward = -20.0
            terminated = True
            
        return self.board.copy(), reward, terminated, False, {}

print("Cella 1: Ambiente Avanzato (SB3 + Nuove Regole) pronto.")
```

```python
# ==============================================================================
# GIOCO VS AI (SB3 VERSION) - CON TIMER REALE
# ==============================================================================

SQUARE_SIZE = 80
WIDTH, HEIGHT = 640, 640
COLORS = {
    'W': (255,255,255), 'B': (0,0,0), 'R': (200,50,50), 
    'Beige': (210,180,140), 'Brown': (139,69,19), 
    'Green': (0,255,0), 'Gold': (255,215,0)
}

def play_vs_sb3():
    model_path = "dama_sb3_ppo.zip"
    if not os.path.exists(model_path):
        print("ERRORE: Esegui prima la Cella 2 per creare il modello.")
        return

    # Carica Modello
    model = PPO.load(model_path)
    
    # Ambiente Grafico
    env = CheckersEnvAdvanced()
    obs, _ = env.reset()
    
    pygame.init()
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Dama SB3 (Tocca a te)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 40)
    
    selected = None
    running = True
    game_over = False
    winner_text = ""
    
    # Variabile per il timer dell'AI
    ai_timer_start = 0 

    while running:
        # 1. GESTIONE EVENTI
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
                pygame.quit()
                return

            # TURNO UMANO (Nero = 1)
            # L'umano può cliccare solo se NON è game over e se NON sta aspettando l'AI
            if event.type == pygame.MOUSEBUTTONDOWN and not game_over and env.current_player == 1:
                pos = pygame.mouse.get_pos()
                r, c = pos[1]//SQUARE_SIZE, pos[0]//SQUARE_SIZE
                
                if env.board[r,c] in [1, 2]:
                    selected = (r,c)
                elif selected:
                    r1, c1 = selected
                    act_idx = r1*512 + c1*64 + r*8 + c
                    
                    if (r1, c1, r, c) in env.possible_moves:
                        obs, _, term, _, _ = env.step(act_idx)
                        selected = None
                        if term:
                            game_over = True
                            winner_text = "HAI VINTO!"
                        
                        # Appena l'umano finisce, resettiamo il timer per l'AI
                        ai_timer_start = 0

        # 2. LOGICA AI (Bianco = 3) CON TIMER
        if env.current_player == 3 and not game_over:
            
            # Se il timer non è ancora partito, fallo partire ora
            if ai_timer_start == 0:
                ai_timer_start = pygame.time.get_ticks()
                pygame.display.set_caption("Dama SB3 (L'IA sta riflettendo...)")
            
            # Controlla se sono passati 800 millisecondi (0.8 secondi)
            current_time = pygame.time.get_ticks()
            if current_time - ai_timer_start > 800:
                
                # --- ORA L'IA GIOCA ---
                action, _ = model.predict(obs, deterministic=True)
                action = int(action)
                
                # Action Masking (Sicurezza)
                r1, c1, r2, c2 = env._decode_action(action)
                if (r1, c1, r2, c2) not in env.possible_moves:
                    print(f"AI ha sbagliato ({r1},{c1}->{r2},{c2}). Correggo...")
                    if env.possible_moves:
                        idx = np.random.randint(len(env.possible_moves))
                        r1, c1, r2, c2 = env.possible_moves[idx]
                        action = r1*512 + c1*64 + r2*8 + c2
                    else:
                        game_over = True
                        winner_text = "AI BLOCCATA (Vinci tu)"
                
                obs, _, term, _, _ = env.step(action)
                
                # Reset
                ai_timer_start = 0 
                pygame.display.set_caption("Dama SB3 (Tocca a te)")
                
                if term:
                    game_over = True
                    winner_text = "AI HA VINTO!"

        # 3. DISEGNO (Avviene SEMPRE, anche mentre l'AI aspetta)
        screen.fill(COLORS['B'])
        for r in range(8):
            for c in range(8):
                col = COLORS['Beige'] if (r+c)%2==0 else COLORS['Brown']
                pygame.draw.rect(screen, col, (c*80, r*80, 80, 80))
                
                # Evidenzia mosse possibili se selezioni
                if selected:
                    sr, sc = selected
                    for m in env.possible_moves:
                        if m[0] == sr and m[1] == sc:
                            pygame.draw.circle(screen, COLORS['Green'], (m[3]*80+40, m[2]*80+40), 10)

                p = env.board[r,c]
                if p != 0:
                    cc = COLORS['R'] if p in [1,2] else COLORS['W']
                    pygame.draw.circle(screen, cc, (c*80+40, r*80+40), 30)
                    if p in [2,4]: # Dama
                        pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 10)
                        pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 32, 2)

        if game_over:
            s = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
            s.fill((0,0,0,200))
            screen.blit(s, (0,0))
            txt = font.render(winner_text, True, COLORS['Green'])
            screen.blit(txt, txt.get_rect(center=(WIDTH/2, HEIGHT/2)))

        pygame.display.flip()
        clock.tick(30)

if __name__ == "__main__":
    play_vs_sb3()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pygame
import time
import os
import torch
import torch.nn as nn
from stable_baselines3 import PPO

# ==============================================================================
# 1. CLASSE DQN (Per caricare il vecchio modello 10k)
# ==============================================================================
DIMENSIONE = 8
ACTION_DIM = DIMENSIONE**4

class DQN(nn.Module):
    def __init__(self):
        super(DQN, self).__init__()
        # Struttura 'net' usata per il modello 10k
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.Flatten(),
            nn.Linear(64*64, 512), nn.ReLU(),
            nn.Linear(512, ACTION_DIM)
        )
    def forward(self, x):
        return self.net(x.unsqueeze(1).float())

# ==============================================================================
# 2. AMBIENTE AVANZATO (Campo di Battaglia)
# ==============================================================================
class CheckersEnvAdvanced(gym.Env):
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        self.action_space = spaces.Discrete(8**4)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()

    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board

    def reset(self, seed=None, options=None):
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}

    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()

    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2

    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                piece = self.board[r, c]
                if piece == self.current_player or piece == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal

    def _get_moves(self, r, c):
        m=[]; piece=self.board[r,c]
        dirs=[(1,-1),(1,1)] if piece==1 else [(-1,-1),(-1,1)]
        if piece in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m

    def _get_captures(self, r, c):
        caps=[]; piece=self.board[r,c]; enemies=[3,4] if piece in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if piece==1 else [(-1,-1),(-1,1)]
        if piece in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd, cd = r+2*dr, c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid = self.board[r+dr, c+dc]
                if mid in enemies and self.board[rd, cd]==0:
                    if piece in [1,3] and mid in [2,4]: continue # Regola Italiana
                    caps.append((r,c,rd,cd))
        return caps

    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; curr_r,curr_c=r,c
        while loop:
            caps = self._get_captures(curr_r, curr_c)
            if not caps: break
            _,_,nr,nc = caps[0]
            self.board[nr,nc]=self.board[curr_r,curr_c]; self.board[curr_r,curr_c]=0
            self.board[(curr_r+nr)//2,(curr_c+nc)//2]=0
            curr_r,curr_c=nr,nc; rew+=2.0
        return curr_r, curr_c, rew

    def step(self, action):
        r1,c1,r2,c2 = self._decode_action(action)
        move=(r1,c1,r2,c2)
        rew=0; term=False
        if move in self.possible_moves:
            p = self.board[r1,c1]
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; rew+=1
                fr,fc,xr = self._execute_capture_chain(r2,c2); r2,c2=fr,fc; rew+=xr
            else: self.board[r2,c2]=p; self.board[r1,c1]=0
            if (self.board[r2,c2]==1 and r2==7) or (self.board[r2,c2]==3 and r2==0):
                self.board[r2,c2]+=1; rew+=5
            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; rew+=10
        else: term=True # Mossa illegale durante test AI vs AI chiude gioco
        return self.board.copy(), rew, term, False, {}

# ==============================================================================
# 3. ARENA: DQN (Old) vs SB3 (New)
# ==============================================================================
FILE_DQN = "dama_10k.pth"
FILE_SB3 = "dama_sb3_ppo.zip"

def ultimate_showdown():
    print(f"--- SCONTRO TRA TITANI ---")
    print(f"NERO (Rosso): DQN Classico (10k episodi)")
    print(f"BIANCO      : SB3 PPO (Nuovo Avanzato)")
    
    # Check Files
    if not os.path.exists(FILE_DQN): print(f"Manca {FILE_DQN}"); return
    if not os.path.exists(FILE_SB3): print(f"Manca {FILE_SB3}"); return

    # 1. Carica DQN (Old School)
    device = torch.device("cpu")
    dqn_model = DQN().to(device)
    try:
        dqn_model.load_state_dict(torch.load(FILE_DQN, map_location=device))
        dqn_model.eval()
        print("DQN Caricato.")
    except:
        print("Errore caricamento DQN. Sicuro sia il formato giusto?")
        return

    # 2. Carica SB3 (New School)
    sb3_model = PPO.load(FILE_SB3)
    print("SB3 PPO Caricato.")

    # Setup Grafico
    pygame.init()
    screen = pygame.display.set_mode((640, 640))
    pygame.display.set_caption("DQN (Rosso) vs SB3 (Bianco)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 40)
    COLORS = {'W':(255,255,255), 'B':(0,0,0), 'R':(200,50,50), 'Beige':(210,180,140), 'Brown':(139,69,19), 'Green':(0,255,0), 'Gold':(255,215,0)}

    env = CheckersEnvAdvanced()
    obs, _ = env.reset()
    
    running = True; game_over = False; winner_text = ""

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT: running = False

        if not game_over:
            pygame.display.flip(); time.sleep(0.8) # Ritardo mossa
            
            legal = env.possible_moves
            if not legal:
                game_over = True
                winner_text = "Vince SB3 (Stallo)" if env.current_player==1 else "Vince DQN (Stallo)"
            else:
                best_idx = -1
                
                # --- TURNO DQN (Nero) ---
                if env.current_player == 1:
                    # DQN Output: Q-Values raw
                    t_obs = torch.tensor(env.board).unsqueeze(0).float().to(device)
                    with torch.no_grad(): q_vals = dqn_model(t_obs).flatten()
                    
                    # Action Masking per DQN
                    best_val = -float('inf')
                    for (r1,c1,r2,c2) in legal:
                        idx = r1*512 + c1*64 + r2*8 + c2
                        if q_vals[idx] > best_val: best_val = q_vals[idx]; best_idx = idx

                # --- TURNO SB3 (Bianco) ---
                else:
                    # SB3 Output: Action index diretto
                    # Nota: SB3 gestisce le sue osservazioni, ma per sicurezza passiamo numpy
                    action, _ = sb3_model.predict(env.board, deterministic=True)
                    act_int = int(action)
                    
                    # Verifica legalità (Action Masking di sicurezza)
                    r1,c1,r2,c2 = env._decode_action(act_int)
                    if (r1,c1,r2,c2) in legal:
                        best_idx = act_int
                    else:
                        # Se SB3 sbaglia, fallback random legale
                        print("SB3 mossa illegale -> Random Fallback")
                        import random
                        r1,c1,r2,c2 = random.choice(legal)
                        best_idx = r1*512 + c1*64 + r2*8 + c2

                # Esegui Mossa
                obs, _, term, _, _ = env.step(best_idx)
                if term:
                    game_over = True
                    winner_text = "Vince SB3 (Bianco)!" if env.current_player==1 else "Vince DQN (Rosso)!"

        # Render
        screen.fill(COLORS['B'])
        for r in range(8):
            for c in range(8):
                col = COLORS['Beige'] if (r+c)%2==0 else COLORS['Brown']
                pygame.draw.rect(screen, col, (c*80, r*80, 80, 80))
                p = env.board[r,c]
                if p!=0:
                    cc = COLORS['R'] if p in [1,2] else COLORS['W']
                    pygame.draw.circle(screen, cc, (c*80+40, r*80+40), 30)
                    if p in [2,4]: pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 10)
        
        if game_over:
            s = pygame.Surface((640,640), pygame.SRCALPHA); s.fill((0,0,0,180)); screen.blit(s,(0,0))
            txt = font.render(winner_text, True, COLORS['Green'])
            screen.blit(txt, txt.get_rect(center=(320,320)))

        pygame.display.flip()
        clock.tick(30)
    pygame.quit()

if __name__ == "__main__":
    ultimate_showdown()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import os
import time
import torch
import torch.nn as nn
from stable_baselines3 import PPO

# ==============================================================================
# 1. DEFINIZIONE AMBIENTE E CLASSI
# ==============================================================================

DIMENSIONE = 8
ACTION_DIM = DIMENSIONE**4

# --- CLASSE DQN (Per leggere il vecchio file .pth) ---
class DQN(nn.Module):
    def __init__(self):
        super(DQN, self).__init__()
        # Struttura usata nel training 10k
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.Flatten(),
            nn.Linear(64*64, 512), nn.ReLU(),
            nn.Linear(512, ACTION_DIM)
        )
    def forward(self, x):
        return self.net(x.unsqueeze(1).float())

# --- CLASSE AMBIENTE AVANZATO ---
class CheckersEnvAdvanced(gym.Env):
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        self.action_space = spaces.Discrete(ACTION_DIM)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()
    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board
    def reset(self, seed=None, options=None):
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}
    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()
    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2
    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                p = self.board[r, c]
                if p == self.current_player or p == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal
    def _get_moves(self, r, c):
        m=[]; p=self.board[r,c]; dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m
    def _get_captures(self, r, c):
        caps=[]; p=self.board[r,c]; en=[3,4] if p in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd,cd=r+2*dr,c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid=self.board[r+dr,c+dc]
                if mid in en and self.board[rd,cd]==0:
                    if p in [1,3] and mid in [2,4]: continue
                    caps.append((r,c,rd,cd))
        return caps
    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; cr,cc=r,c
        while loop:
            caps=self._get_captures(cr,cc)
            if not caps: break
            _,_,nr,nc=caps[0]
            self.board[nr,nc]=self.board[cr,cc]; self.board[cr,cc]=0
            self.board[(cr+nr)//2,(cc+nc)//2]=0
            cr,cc=nr,nc; rew+=2.0
        return cr,cc,rew
    def step(self, action):
        r1,c1,r2,c2=self._decode_action(action); move=(r1,c1,r2,c2); rew=0; term=False
        if move in self.possible_moves:
            p=self.board[r1,c1]
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; rew+=1
                fr,fc,xr=self._execute_capture_chain(r2,c2); r2,c2=fr,fc; rew+=xr
            else: self.board[r2,c2]=p; self.board[r1,c1]=0
            if (self.board[r2,c2]==1 and r2==7) or (self.board[r2,c2]==3 and r2==0): self.board[r2,c2]+=1; rew+=5
            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; rew+=10
        else: rew=-20; term=True
        return self.board.copy(), rew, term, False, {}

# ==============================================================================
# 2. SIMULAZIONE IBRIDA: PyTorch (.pth) vs SB3 (.zip)
# ==============================================================================

FILE_100K = "dama_100k_ppo.zip"

# Cerca il file da 50k (o usa quello sb3 standard se non trovi l'altro)
FILE_50K = "dama_50k.zip"
if not os.path.exists(FILE_50K) and os.path.exists("dama_sb3_ppo.zip"):
    FILE_50K = "dama_sb3_ppo.zip"

def run_hybrid_simulation():
    print(f"--- SCONTRO GENERAZIONALE ---")
    
    # 1. Carica il NERO (Vecchio PyTorch)
    if not os.path.exists(FILE_PTH_10K):
        print(f"ERRORE: Manca {FILE_PTH_10K}"); return
    
    device = torch.device("cpu")
    model_old = DQN().to(device)
    try:
        model_old.load_state_dict(torch.load(FILE_PTH_10K, map_location=device))
        model_old.eval()
        print(f"NERO (P1): Caricato '{FILE_PTH_10K}' (PyTorch)")
    except Exception as e:
        print(f"Errore caricamento .pth: {e}"); return

    # 2. Carica il BIANCO (Nuovo SB3)
    if not os.path.exists(FILE_ZIP_50K):
        print(f"ERRORE: Manca {FILE_ZIP_50K}"); return
    
    try:
        model_new = PPO.load(FILE_ZIP_50K)
        print(f"BIANCO (P2): Caricato '{FILE_ZIP_50K}' (Stable Baselines)")
    except Exception as e:
        print(f"Errore caricamento .zip: {e}"); return
    
    # 3. Avvia Simulazione
    env = CheckersEnvAdvanced()
    
    total_games = 1000
    wins_old = 0
    wins_new = 0
    draws = 0
    max_moves = 100
    
    start_time = time.time()
    print(f"\nSimulazione di {total_games} partite in corso...")
    
    for i in range(total_games):
        obs, _ = env.reset()
        moves = 0
        game_over = False
        
        while not game_over:
            moves += 1
            if moves >= max_moves:
                game_over = True; draws += 1; break
            
            # --- TURNO ---
            if env.current_player == 1: 
                # NERO -> Usa PyTorch (DQN)
                t_obs = torch.tensor(obs).unsqueeze(0).float().to(device)
                with torch.no_grad():
                    q_vals = model_old(t_obs).flatten()
                
                # Sceglie la mossa migliore tra quelle legali (Action Masking)
                best_idx = -1; best_val = -float('inf')
                # Fallback se non ci sono mosse (non dovrebbe succedere se game_over=False)
                if not env.possible_moves: game_over=True; wins_new+=1; break
                
                for (r1,c1,r2,c2) in env.possible_moves:
                    idx = r1*512 + c1*64 + r2*8 + c2
                    if q_vals[idx] > best_val: best_val = q_vals[idx]; best_idx = idx
                action = best_idx

            else: 
                # BIANCO -> Usa SB3 (PPO)
                action, _ = model_new.predict(obs, deterministic=True)
                action = int(action)
                # Verifica legalità SB3
                r1,c1,r2,c2 = env._decode_action(action)
                if (r1,c1,r2,c2) not in env.possible_moves:
                    if env.possible_moves:
                        # Fallback random legale
                        import random
                        r1,c1,r2,c2 = random.choice(env.possible_moves)
                        action = r1*512 + c1*64 + r2*8 + c2
                    else:
                        game_over=True; wins_old+=1; break

            # Esegui step
            obs, _, term, _, _ = env.step(action)
            
            if term:
                game_over = True
                if env.current_player == 3: wins_old += 1 # Ha vinto Nero (ha mosso per ultimo ed è diventato turno Bianco)
                else: wins_new += 1 # Ha vinto Bianco

        if (i+1) % 100 == 0:
            print(f"Partita {i+1} completata...")

    end_time = time.time()
    
    print("\n" + "="*40)
    print(" RISULTATI FINALI")
    print("="*40)
    print(f"NERO (10k Vecchio): {wins_old} vittorie ({wins_old/total_games*100:.1f}%)")
    print(f"BIANCO (50k Nuovo): {wins_new} vittorie ({wins_new/total_games*100:.1f}%)")
    print(f"Pareggi: {draws} ({draws/total_games*100:.1f}%)")
    print(f"Tempo: {end_time - start_time:.2f}s")

if __name__ == "__main__":
    run_hybrid_simulation()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pygame
import time
import os
from stable_baselines3 import PPO

# ==============================================================================
# 1. AMBIENTE AVANZATO (Necessario)
# ==============================================================================
class CheckersEnvAdvanced(gym.Env):
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        self.action_space = spaces.Discrete(8**4)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()
    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board
    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}
    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()
    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2
    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                p = self.board[r, c]
                if p == self.current_player or p == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal
    def _get_moves(self, r, c):
        m=[]; p=self.board[r,c]; dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m
    def _get_captures(self, r, c):
        caps=[]; p=self.board[r,c]; en=[3,4] if p in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd,cd=r+2*dr,c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid=self.board[r+dr,c+dc]
                if mid in en and self.board[rd,cd]==0:
                    if p in [1,3] and mid in [2,4]: continue
                    caps.append((r,c,rd,cd))
        return caps
    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; cr,cc=r,c
        while loop:
            caps=self._get_captures(cr,cc)
            if not caps: break
            _,_,nr,nc=caps[0]
            self.board[nr,nc]=self.board[cr,cc]; self.board[cr,cc]=0
            self.board[(cr+nr)//2,(cc+nc)//2]=0
            cr,cc=nr,nc; rew+=2.0
        return cr,cc,rew
    def step(self, action):
        r1,c1,r2,c2=self._decode_action(action); move=(r1,c1,r2,c2); rew=0; term=False
        if move in self.possible_moves:
            p=self.board[r1,c1]
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; rew+=1
                fr,fc,xr=self._execute_capture_chain(r2,c2); r2,c2=fr,fc; rew+=xr
            else: self.board[r2,c2]=p; self.board[r1,c1]=0
            if (self.board[r2,c2]==1 and r2==7) or (self.board[r2,c2]==3 and r2==0): self.board[r2,c2]+=1; rew+=5
            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; rew+=10
        else: rew=-20; term=True
        return self.board.copy(), rew, term, False, {}

# ==============================================================================
# 2. SCONTRO VISIVO: 100k vs 50k
# ==============================================================================

# Definisci i nomi dei file
FILE_100K = "dama_100k_ppo.zip"

# Cerca il file da 50k (o usa quello sb3 standard se non trovi l'altro)
FILE_50K = "dama_50k.zip"
if not os.path.exists(FILE_50K) and os.path.exists("dama_sb3_ppo.zip"):
    FILE_50K = "dama_sb3_ppo.zip"

SQUARE_SIZE = 80
WIDTH, HEIGHT = 640, 640
COLORS = {'W':(255,255,255), 'B':(0,0,0), 'R':(200,50,50), 'Beige':(210,180,140), 'Brown':(139,69,19), 'Green':(0,255,0), 'Gold':(255,215,0)}

def showdown_100k_vs_50k():
    print("--- SCONTRO TITANI ---")
    
    # Verifica file
    if not os.path.exists(FILE_100K): print(f"Manca {FILE_100K}"); return
    if not os.path.exists(FILE_50K): print(f"Manca {FILE_50K}"); return

    print(f"NERO (P1): 100k ({FILE_100K})")
    print(f"BIANCO (P2): 50k ({FILE_50K})")

    # Carica Modelli
    model_100k = PPO.load(FILE_100K)
    model_50k = PPO.load(FILE_50K)

    # Inizializza Pygame
    pygame.init()
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption(f"100k (Nero) vs 50k (Bianco)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 40)
    
    env = CheckersEnvAdvanced()
    obs, _ = env.reset()
    
    running = True; game_over = False; winner_text = ""
    
    # Delay per vedere le mosse (0.8 secondi)
    MOVE_DELAY = 0.8 

    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT: running = False

        if not game_over:
            # Renderizza PRIMA di aspettare
            screen.fill(COLORS['B'])
            for r in range(8):
                for c in range(8):
                    col = COLORS['Beige'] if (r+c)%2==0 else COLORS['Brown']
                    pygame.draw.rect(screen, col, (c*80, r*80, 80, 80))
                    p = env.board[r,c]
                    if p!=0:
                        cc = COLORS['R'] if p in [1,2] else COLORS['W']
                        pygame.draw.circle(screen, cc, (c*80+40, r*80+40), 30)
                        if p in [2,4]: pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 10)
            pygame.display.flip()
            
            # Aspetta
            time.sleep(MOVE_DELAY)
            
            legal = env.possible_moves
            if not legal:
                game_over = True
                winner_text = "Vince BIANCO (Stallo)" if env.current_player==1 else "Vince NERO (Stallo)"
            else:
                # SCEGLI AZIONE
                if env.current_player == 1:
                    # NERO -> 100k
                    action, _ = model_100k.predict(obs, deterministic=True)
                else:
                    # BIANCO -> 50k
                    action, _ = model_50k.predict(obs, deterministic=True)
                
                action = int(action)
                
                # Action Masking (Fallback)
                r1,c1,r2,c2 = env._decode_action(action)
                if (r1,c1,r2,c2) not in legal:
                    if legal:
                        import random
                        idx = random.randint(0, len(legal)-1)
                        r1,c1,r2,c2 = legal[idx]
                        action = r1*512 + c1*64 + r2*8 + c2
                    else:
                        game_over = True # Dovrebbe essere gestito sopra ma per sicurezza
                
                obs, _, term, _, _ = env.step(action)
                
                if term:
                    game_over = True
                    winner_text = "Vince 100k (Nero)!" if env.current_player==3 else "Vince 50k (Bianco)!"

        # Disegno Finale (se game over)
        if game_over:
            s = pygame.Surface((WIDTH,HEIGHT), pygame.SRCALPHA); s.fill((0,0,0,180)); screen.blit(s,(0,0))
            txt = font.render(winner_text, True, COLORS['Green'])
            screen.blit(txt, txt.get_rect(center=(WIDTH/2, HEIGHT/2)))
            pygame.display.flip()

        clock.tick(30)
    pygame.quit()

if __name__ == "__main__":
    showdown_100k_vs_50k()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import os
import time
import torch
import torch.nn as nn
from stable_baselines3 import PPO

# ==============================================================================
# 1. CONFIGURAZIONE NOMI FILE (CONTROLLA QUESTI NOMI!)
# ==============================================================================

# Il file del modello vecchio (PyTorch)
FILE_OLD_PYTORCH = "dama_10k.pth"  
# Se non hai 'dama_10k.pth', prova "dama_dqn_final.pth"

# Il file del modello nuovo (Stable Baselines)
FILE_NEW_SB3 = "dama_50k.zip"      
# Se non hai 'dama_50k.zip', il codice cercherà "dama_sb3_ppo.zip" in automatico

# ==============================================================================
# 2. DEFINIZIONE CLASSI E AMBIENTE
# ==============================================================================

DIMENSIONE = 8
ACTION_DIM = DIMENSIONE**4

# --- CLASSE DQN (Per leggere il vecchio file .pth) ---
class DQN(nn.Module):
    def __init__(self):
        super(DQN, self).__init__()
        # Struttura usata nel training 10k
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.Flatten(),
            nn.Linear(64*64, 512), nn.ReLU(),
            nn.Linear(512, ACTION_DIM)
        )
    def forward(self, x):
        return self.net(x.unsqueeze(1).float())

# --- CLASSE AMBIENTE AVANZATO ---
class CheckersEnvAdvanced(gym.Env):
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        self.action_space = spaces.Discrete(ACTION_DIM)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()
    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board
    def reset(self, seed=None, options=None):
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}
    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()
    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2
    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                p = self.board[r, c]
                if p == self.current_player or p == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal
    def _get_moves(self, r, c):
        m=[]; p=self.board[r,c]; dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m
    def _get_captures(self, r, c):
        caps=[]; p=self.board[r,c]; en=[3,4] if p in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd,cd=r+2*dr,c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid=self.board[r+dr,c+dc]
                if mid in en and self.board[rd,cd]==0:
                    if p in [1,3] and mid in [2,4]: continue
                    caps.append((r,c,rd,cd))
        return caps
    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; cr,cc=r,c
        while loop:
            caps=self._get_captures(cr,cc)
            if not caps: break
            _,_,nr,nc=caps[0]
            self.board[nr,nc]=self.board[cr,cc]; self.board[cr,cc]=0
            self.board[(cr+nr)//2,(cc+nc)//2]=0
            cr,cc=nr,nc; rew+=2.0
        return cr,cc,rew
    def step(self, action):
        r1,c1,r2,c2=self._decode_action(action); move=(r1,c1,r2,c2); rew=0; term=False
        if move in self.possible_moves:
            p=self.board[r1,c1]
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; rew+=1
                fr,fc,xr=self._execute_capture_chain(r2,c2); r2,c2=fr,fc; rew+=xr
            else: self.board[r2,c2]=p; self.board[r1,c1]=0
            if (self.board[r2,c2]==1 and r2==7) or (self.board[r2,c2]==3 and r2==0): self.board[r2,c2]+=1; rew+=5
            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; rew+=10
        else: rew=-20; term=True
        return self.board.copy(), rew, term, False, {}

# ==============================================================================
# 3. SIMULAZIONE IBRIDA: PyTorch vs SB3 (CORRETTA)
# ==============================================================================

def run_hybrid_simulation():
    # Gestione Fallback nomi file
    global FILE_NEW_SB3
    if not os.path.exists(FILE_NEW_SB3) and os.path.exists("dama_sb3_ppo.zip"):
        FILE_NEW_SB3 = "dama_sb3_ppo.zip"

    print(f"--- SCONTRO GENERAZIONALE ---")
    print(f"File PyTorch (Old): {FILE_OLD_PYTORCH}")
    print(f"File SB3 (New):     {FILE_NEW_SB3}")
    
    # 1. Carica il NERO (Vecchio PyTorch)
    if not os.path.exists(FILE_OLD_PYTORCH):
        print(f"ERRORE: Manca {FILE_OLD_PYTORCH}. Assicurati di avere il file .pth"); return
    
    device = torch.device("cpu")
    model_old = DQN().to(device)
    try:
        model_old.load_state_dict(torch.load(FILE_OLD_PYTORCH, map_location=device))
        model_old.eval()
        print(f"NERO (P1): Caricato con successo.")
    except Exception as e:
        print(f"Errore caricamento .pth: {e}"); return

    # 2. Carica il BIANCO (Nuovo SB3)
    if not os.path.exists(FILE_NEW_SB3):
        print(f"ERRORE: Manca {FILE_NEW_SB3}. Assicurati di avere il file .zip"); return
    
    try:
        model_new = PPO.load(FILE_NEW_SB3)
        print(f"BIANCO (P2): Caricato con successo.")
    except Exception as e:
        print(f"Errore caricamento .zip: {e}"); return
    
    # 3. Avvia Simulazione
    env = CheckersEnvAdvanced()
    
    total_games = 1000
    wins_old = 0
    wins_new = 0
    draws = 0
    max_moves = 100
    
    start_time = time.time()
    print(f"\nSimulazione di {total_games} partite in corso...")
    
    for i in range(total_games):
        obs, _ = env.reset()
        moves = 0
        game_over = False
        
        while not game_over:
            moves += 1
            if moves >= max_moves:
                game_over = True; draws += 1; break
            
            # --- TURNO ---
            if env.current_player == 1: 
                # NERO -> Usa PyTorch (DQN)
                t_obs = torch.tensor(obs).unsqueeze(0).float().to(device)
                with torch.no_grad():
                    q_vals = model_old(t_obs).flatten()
                
                # Action Masking per PyTorch
                best_idx = -1; best_val = -float('inf')
                if not env.possible_moves: game_over=True; wins_new+=1; break
                
                for (r1,c1,r2,c2) in env.possible_moves:
                    idx = r1*512 + c1*64 + r2*8 + c2
                    if q_vals[idx] > best_val: best_val = q_vals[idx]; best_idx = idx
                action = best_idx

            else: 
                # BIANCO -> Usa SB3 (PPO)
                action, _ = model_new.predict(obs, deterministic=True)
                action = int(action)
                
                # Action Masking / Fallback per SB3
                r1,c1,r2,c2 = env._decode_action(action)
                if (r1,c1,r2,c2) not in env.possible_moves:
                    if env.possible_moves:
                        import random
                        r1,c1,r2,c2 = random.choice(env.possible_moves)
                        action = r1*512 + c1*64 + r2*8 + c2
                    else:
                        game_over=True; wins_old+=1; break

            # Esegui step
            obs, _, term, _, _ = env.step(action)
            
            if term:
                game_over = True
                if env.current_player == 3: wins_old += 1 # Ha vinto Nero (P1)
                else: wins_new += 1 # Ha vinto Bianco (P2)

        if (i+1) % 100 == 0:
            print(f"Partita {i+1} completata...")

    end_time = time.time()
    
    print("\n" + "="*40)
    print(" RISULTATI FINALI")
    print("="*40)
    print(f"NERO (PyTorch Old): {wins_old} vittorie ({wins_old/total_games*100:.1f}%)")
    print(f"BIANCO (SB3 New):   {wins_new} vittorie ({wins_new/total_games*100:.1f}%)")
    print(f"Pareggi: {draws} ({draws/total_games*100:.1f}%)")
    print(f"Tempo: {end_time - start_time:.2f}s")

if __name__ == "__main__":
    run_hybrid_simulation()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import pygame
import time
import os
import torch
import torch.nn as nn
from stable_baselines3 import PPO

# ==============================================================================
# 1. CONFIGURAZIONE NOMI FILE (Verifica che i nomi siano corretti!)
# ==============================================================================
# Assumo questi nomi basandomi sulle chat precedenti. 
# Se i tuoi file hanno nomi diversi, modificali qui.

FILE_1K   = "dama_dqn_final.pth"   # Livello Principiante (o dama_1k.pth)
FILE_10K  = "dama_10k.pth"         # Livello Intermedio
FILE_100K = "dama_100k_ppo.zip"    # Livello Maestro (o dama_sb3_ppo.zip)

# Fallback se non hai fatto il 100k ma hai il 50k
if not os.path.exists(FILE_100K) and os.path.exists("dama_50k.zip"):
    FILE_100K = "dama_50k.zip"

# ==============================================================================
# 2. DEFINIZIONE CLASSI (DQN, Ambiente)
# ==============================================================================
DIMENSIONE = 8
ACTION_DIM = DIMENSIONE**4

# --- CLASSE DQN (Per i vecchi modelli .pth) ---
class DQN(nn.Module):
    def __init__(self):
        super(DQN, self).__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.Flatten(),
            nn.Linear(64*64, 512), nn.ReLU(),
            nn.Linear(512, ACTION_DIM)
        )
    def forward(self, x):
        return self.net(x.unsqueeze(1).float())

# --- CLASSE AMBIENTE AVANZATO ---
class CheckersEnvAdvanced(gym.Env):
    def __init__(self):
        super(CheckersEnvAdvanced, self).__init__()
        self.action_space = spaces.Discrete(ACTION_DIM)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()
    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board
    def reset(self, seed=None, options=None):
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}
    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()
    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2
    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                p = self.board[r, c]
                if p == self.current_player or p == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal
    def _get_moves(self, r, c):
        m=[]; p=self.board[r,c]; dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m
    def _get_captures(self, r, c):
        caps=[]; p=self.board[r,c]; en=[3,4] if p in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd,cd=r+2*dr,c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid=self.board[r+dr,c+dc]
                if mid in en and self.board[rd,cd]==0:
                    if p in [1,3] and mid in [2,4]: continue
                    caps.append((r,c,rd,cd))
        return caps
    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; cr,cc=r,c
        while loop:
            caps=self._get_captures(cr,cc)
            if not caps: break
            _,_,nr,nc=caps[0]
            self.board[nr,nc]=self.board[cr,cc]; self.board[cr,cc]=0
            self.board[(cr+nr)//2,(cc+nc)//2]=0
            cr,cc=nr,nc; rew+=2.0
        return cr,cc,rew
    def step(self, action):
        r1,c1,r2,c2=self._decode_action(action); move=(r1,c1,r2,c2); rew=0; term=False
        if move in self.possible_moves:
            p=self.board[r1,c1]
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; rew+=1
                fr,fc,xr=self._execute_capture_chain(r2,c2); r2,c2=fr,fc; rew+=xr
            else: self.board[r2,c2]=p; self.board[r1,c1]=0
            if (self.board[r2,c2]==1 and r2==7) or (self.board[r2,c2]==3 and r2==0): self.board[r2,c2]+=1; rew+=5
            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; rew+=10
        else: rew=-20; term=True
        return self.board.copy(), rew, term, False, {}

# ==============================================================================
# 3. INTERFACCIA E LOGICA GIOCO
# ==============================================================================

SQUARE_SIZE = 80
WIDTH, HEIGHT = 640, 640
COLORS = {
    'W': (255,255,255), 'B': (0,0,0), 'R': (200,50,50), 
    'Beige': (210,180,140), 'Brown': (139,69,19), 
    'Green': (0,255,0), 'Gold': (255,215,0), 'Blue': (50, 50, 200)
}

# --- WRAPPER PER UNIFICARE I MODELLI ---
# Questa classe nasconde la complessità: prende un modello (che sia torch o sb3)
# e restituisce sempre un'azione intera valida.
class AI_Player:
    def __init__(self, model_path, model_type):
        self.model_type = model_type
        self.device = torch.device("cpu")
        
        if model_type == "sb3":
            print(f"Caricamento PPO SB3: {model_path}")
            self.model = PPO.load(model_path)
        elif model_type == "torch":
            print(f"Caricamento DQN PyTorch: {model_path}")
            self.model = DQN().to(self.device)
            self.model.load_state_dict(torch.load(model_path, map_location=self.device))
            self.model.eval()

    def get_action(self, obs, legal_moves):
        # 1. PPO (Stable Baselines)
        if self.model_type == "sb3":
            action, _ = self.model.predict(obs, deterministic=True)
            action = int(action)
            # Verifica se è legale, altrimenti random fallback
            # (SB3 gestisce internamente molto, ma l'Action Masking è sicuro)
            return action

        # 2. DQN (PyTorch)
        elif self.model_type == "torch":
            t_obs = torch.tensor(obs).unsqueeze(0).float().to(self.device)
            with torch.no_grad():
                q_vals = self.model(t_obs).flatten()
            
            # Action Masking Obbligatorio per DQN
            best_idx = -1
            best_val = -float('inf')
            
            # Se non ci sono mosse, ritorna 0 (tanto il gioco finirà)
            if not legal_moves: return 0 
            
            for (r1,c1,r2,c2) in legal_moves:
                idx = r1*512 + c1*64 + r2*8 + c2
                if q_vals[idx] > best_val:
                    best_val = q_vals[idx]
                    best_idx = idx
            return best_idx

def main_game():
    pygame.init()
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Dama AI Championship")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 50)
    small_font = pygame.font.SysFont(None, 30)

    # --- FASE 1: MENU DI SELEZIONE ---
    selected_ai = None
    menu_running = True
    
    while menu_running:
        screen.fill(COLORS['B'])
        
        # Testi Menu
        title = font.render("SCEGLI DIFFICOLTA'", True, COLORS['Green'])
        txt1 = font.render("1. Principiante (1k)", True, COLORS['W'])
        txt2 = font.render("2. Intermedio (10k)", True, COLORS['W'])
        txt3 = font.render("3. Maestro (100k)", True, COLORS['Gold'])
        info = small_font.render("Premi 1, 2 o 3 sulla tastiera", True, (150,150,150))
        
        screen.blit(title, (WIDTH//2 - title.get_width()//2, 100))
        screen.blit(txt1, (WIDTH//2 - txt1.get_width()//2, 250))
        screen.blit(txt2, (WIDTH//2 - txt2.get_width()//2, 320))
        screen.blit(txt3, (WIDTH//2 - txt3.get_width()//2, 390))
        screen.blit(info, (WIDTH//2 - info.get_width()//2, 550))
        
        pygame.display.flip()
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); return
            
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
                    if os.path.exists(FILE_1K):
                        selected_ai = AI_Player(FILE_1K, "torch")
                        pygame.display.set_caption(f"Tu vs Principiante (1k)")
                        menu_running = False
                    else: print(f"File {FILE_1K} non trovato!")
                
                elif event.key == pygame.K_2:
                    if os.path.exists(FILE_10K):
                        selected_ai = AI_Player(FILE_10K, "torch")
                        pygame.display.set_caption(f"Tu vs Intermedio (10k)")
                        menu_running = False
                    else: print(f"File {FILE_10K} non trovato!")
                    
                elif event.key == pygame.K_3:
                    if os.path.exists(FILE_100K):
                        selected_ai = AI_Player(FILE_100K, "sb3")
                        pygame.display.set_caption(f"Tu vs Maestro (100k)")
                        menu_running = False
                    else: print(f"File {FILE_100K} non trovato!")

    # --- FASE 2: LOOP DI GIOCO ---
    env = CheckersEnvAdvanced()
    obs, _ = env.reset()
    
    selected_piece = None
    game_running = True
    game_over = False
    winner_text = ""
    
    # Timer AI
    ai_timer_start = 0
    AI_THINK_TIME = 800 # millisecondi

    while game_running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_running = False
                pygame.quit(); return

            # INPUT UMANO (Solo se turno 1 e non game over)
            if event.type == pygame.MOUSEBUTTONDOWN and not game_over and env.current_player == 1:
                pos = pygame.mouse.get_pos()
                r, c = pos[1]//SQUARE_SIZE, pos[0]//SQUARE_SIZE
                
                # Selezione
                if env.board[r,c] in [1, 2]:
                    selected_piece = (r,c)
                # Movimento
                elif selected_piece:
                    r1, c1 = selected_piece
                    act_idx = r1*512 + c1*64 + r*8 + c
                    
                    if (r1,c1,r,c) in env.possible_moves:
                        obs, _, term, _, _ = env.step(act_idx)
                        selected_piece = None
                        if term:
                            game_over = True
                            winner_text = "HAI VINTO!"
                        else:
                             # Avvia timer AI
                             ai_timer_start = pygame.time.get_ticks()

        # LOGICA AI (Bianco = 3)
        if env.current_player == 3 and not game_over:
            # Controllo Tempo
            now = pygame.time.get_ticks()
            if now - ai_timer_start > AI_THINK_TIME:
                
                # Ottieni mossa dall'AI selezionata
                action = selected_ai.get_action(obs, env.possible_moves)
                
                # Safety Check (Fallback)
                r1, c1, r2, c2 = env._decode_action(action)
                if (r1, c1, r2, c2) not in env.possible_moves:
                    print("AI ha scelto mossa illegale. Fallback random.")
                    if env.possible_moves:
                        import random
                        r1,c1,r2,c2 = random.choice(env.possible_moves)
                        action = r1*512 + c1*64 + r2*8 + c2
                    else:
                        game_over = True
                        winner_text = "AI BLOCCATA (Vinci tu)"
                
                obs, _, term, _, _ = env.step(action)
                
                if term:
                    game_over = True
                    winner_text = "L'AI HA VINTO!"

        # DISEGNO
        screen.fill(COLORS['B'])
        for r in range(8):
            for c in range(8):
                col = COLORS['Beige'] if (r+c)%2==0 else COLORS['Brown']
                pygame.draw.rect(screen, col, (c*80, r*80, 80, 80))
                
                # Suggerimenti
                if selected_piece:
                    sr, sc = selected_piece
                    for m in env.possible_moves:
                        if m[0] == sr and m[1] == sc:
                            pygame.draw.circle(screen, COLORS['Green'], (m[3]*80+40, m[2]*80+40), 10)

                p = env.board[r,c]
                if p != 0:
                    cc = COLORS['R'] if p in [1,2] else COLORS['W']
                    pygame.draw.circle(screen, cc, (c*80+40, r*80+40), 30)
                    if p in [2,4]: 
                         pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 10)
                         pygame.draw.circle(screen, COLORS['Gold'], (c*80+40, r*80+40), 32, 2)

        if game_over:
            s = pygame.Surface((WIDTH,HEIGHT), pygame.SRCALPHA); s.fill((0,0,0,200)); screen.blit(s,(0,0))
            txt = font.render(winner_text, True, COLORS['Green'])
            screen.blit(txt, txt.get_rect(center=(WIDTH/2, HEIGHT/2)))

        pygame.display.flip()
        clock.tick(30)

if __name__ == "__main__":
    main_game()
```

```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import os
import torch
import torch.nn as nn
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from typing import Callable

# ==============================================================================
# 1. FUNZIONE PER ALFA DINAMICO (SCHEDULER)
# ==============================================================================
def linear_schedule(initial_value: float) -> Callable[[float], float]:
    """
    Funzione che riduce il learning rate linearmente.
    :param initial_value: Il valore di partenza (Alto)
    :return: funzione che calcola il LR corrente
    """
    def func(progress_remaining: float) -> float:
        # progress_remaining va da 1.0 (inizio) a 0.0 (fine)
        return progress_remaining * initial_value
    return func

# ==============================================================================
# 2. AMBIENTE GRANDMASTER (Strategico)
# ==============================================================================
class CheckersEnvGrandmaster(gym.Env):
    def __init__(self):
        super(CheckersEnvGrandmaster, self).__init__()
        self.action_space = spaces.Discrete(8**4)
        self.observation_space = spaces.Box(low=0, high=4, shape=(8, 8), dtype=np.uint8)
        self.player_map = {1: 2, 3: 4}
        self.board = self._initialize_board()
        self.current_player = 1
        self.possible_moves = []
        self._update_moves()

    def _initialize_board(self):
        board = np.zeros((8, 8), dtype=np.uint8)
        for r in range(3):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 1
        for r in range(5, 8):
            for c in range(8):
                if (r + c) % 2 == 1: board[r, c] = 3
        return board

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.board = self._initialize_board()
        self.current_player = 1
        self._update_moves()
        return self.board.copy(), {}

    def _update_moves(self):
        self.possible_moves = self._find_all_legal_moves()

    def _decode_action(self, action_int):
        r1 = action_int // 512; c1 = (action_int // 64) % 8
        r2 = (action_int // 8) % 8; c2 = action_int % 8
        return r1, c1, r2, c2

    def _find_all_legal_moves(self):
        captures, normal = [], []
        king = self.player_map.get(self.current_player, 0)
        for r in range(8):
            for c in range(8):
                p = self.board[r, c]
                if p == self.current_player or p == king:
                    captures.extend(self._get_captures(r, c))
                    if not captures: normal.extend(self._get_moves(r, c))
        return captures if captures else normal

    def _get_moves(self, r, c):
        m=[]; p=self.board[r,c]; dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            if 0<=r+dr<8 and 0<=c+dc<8 and self.board[r+dr,c+dc]==0: m.append((r,c,r+dr,c+dc))
        return m

    def _get_captures(self, r, c):
        caps=[]; p=self.board[r,c]; en=[3,4] if p in [1,2] else [1,2]
        dirs=[(1,-1),(1,1)] if p==1 else [(-1,-1),(-1,1)]
        if p in [2,4]: dirs=[(1,-1),(1,1),(-1,-1),(-1,1)]
        for dr,dc in dirs:
            rd,cd=r+2*dr,c+2*dc
            if 0<=rd<8 and 0<=cd<8:
                mid=self.board[r+dr,c+dc]
                if mid in en and self.board[rd,cd]==0:
                    if p in [1,3] and mid in [2,4]: continue
                    caps.append((r,c,rd,cd))
        return caps

    def _execute_capture_chain(self, r, c):
        rew=0; loop=True; cr,cc=r,c
        while loop:
            caps=self._get_captures(cr,cc)
            if not caps: break
            _,_,nr,nc=caps[0]
            self.board[nr,nc]=self.board[cr,cc]; self.board[cr,cc]=0
            self.board[(cr+nr)//2,(cc+nc)//2]=0
            cr,cc=nr,nc; rew+=2.0
        return cr,cc,rew

    def step(self, action):
        r1,c1,r2,c2 = self._decode_action(action)
        move=(r1,c1,r2,c2)
        reward=0; term=False
        
        if move in self.possible_moves:
            p = self.board[r1,c1]
            
            # --- STRATEGIA POSIZIONALE (Grandmaster) ---
            # 1. Premia controllo del centro (Righe 3,4 Colonne 2-5)
            if r2 in [3, 4] and c2 in [2, 3, 4, 5]:
                reward += 0.3 # Strategia chiave nella dama
            
            # 2. Premia la creazione di Dame
            is_promo = False
            if (p==1 and r2==7) or (p==3 and r2==0):
                is_promo = True

            # Esecuzione
            if abs(r2-r1)==2:
                self.board[r2,c2]=p; self.board[r1,c1]=0; self.board[(r1+r2)//2,(c1+c2)//2]=0; reward+=1.0
                fr,fc,xr = self._execute_capture_chain(r2,c2); r2,c2=fr,fc; reward+=xr
            else:
                self.board[r2,c2]=p; self.board[r1,c1]=0
            
            if is_promo:
                self.board[r2,c2] += 1; reward += 5.0 
            
            # 3. Penalità per stallo (muovere avanti e indietro inutilmente)
            # (Impliciata nel fatto che il gioco deve finire per avere il big reward)

            self.current_player=3 if self.current_player==1 else 1
            self._update_moves()
            if not self.possible_moves: term=True; reward+=20.0 # Vittoria schiacciante
        else:
            reward = -50.0; term=True # Mossa illegale (punizione severa)
            
        return self.board.copy(), reward, term, False, {}

# ==============================================================================
# 3. CONFIGURAZIONE TRAINING HARDCORE
# ==============================================================================

FILENAME_HARD = "dama_grandmaster_1m"

env = DummyVecEnv([lambda: CheckersEnvGrandmaster()])

# Rete Neurale Molto Profonda (Deep Network)
# pi = Policy (Decisione), vf = Value Function (Valutazione)
# [512, 512, 256] significa 3 strati enormi
policy_kwargs = dict(
    activation_fn=torch.nn.ReLU,
    net_arch=dict(pi=[512, 512, 256], vf=[512, 512, 256])
)

# Configurazione PPO "Aggressiva" ma Stabile
model = PPO(
    "MlpPolicy", 
    env, 
    verbose=1, 
    
    # QUI C'È IL TUO ALFA ELEVATO, MA DINAMICO
    # Parte da 0.001 (molto alto per PPO) e scende a 0
    learning_rate=linear_schedule(0.001), 
    
    n_steps=2048,               # Buffer di esperienza
    batch_size=256,             # Batch grande per stabilizzare l'alfa alto
    ent_coef=0.01,              # Entropia per forzare esplorazione creativa
    gamma=0.995,                # Guarda molto lontano nel futuro
    policy_kwargs=policy_kwargs # Usa il cervello grande
)

print(f"--- TRAINING GRANDMASTER (Alfa Alto Decrescente + Deep Net) ---")
print("Obiettivo: 1.000.000 steps.")
print("Nota: Il learning rate partirà alto (0.001) per imparare veloce e scenderà per raffinare.")

# Addestramento
model.learn(total_timesteps=1000000)

model.save(FILENAME_HARD)
print(f"\n>>> SALVATO: {FILENAME_HARD}.zip")
```
